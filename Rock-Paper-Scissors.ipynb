```python
import random
from PIL import Image
from IPython.display import display
```

```python
def get_user_choice():
    user_choice = input("Enter your choice (rock, paper, scissors): ").lower()
    while user_choice not in ['rock','paper', 'scissors']:
        user_choice = input("Invalid choice. Please enter rock, paper, or scissors: ").lower()
    return user_choice

def get_computer_choice():
    return random.choice(['rock', 'paper', 'scissors'])

def determine_winner(Player1, Player2):
    if Player1 == Player2:
        return "It's a tie!"
    elif (Player1 == 'rock' and Player2 == 'scissors') or \
         (Player1 == 'paper' and Player2 == 'rock') or \
         (Player1 == 'scissors' and Player2 == 'paper'):
        return "You win!"
    else:
        return "You lose!"
```

```python
Player1 = random.choice(['rock', 'paper', 'scissors'])
Player2 = random.choice(['rock', 'paper', 'scissors'])
print("First Player",Player1,"Second Player",Player2)
determine_winner(Player1, Player2)
```

```python
def show(choice):

    if choice.lower() == 'rock':
        image = Image.open('/content/Rock.png')
    elif choice.lower() == 'paper':
        image = Image.open('/content/Paper.png')
    elif choice.lower() == 'scissors':
        image = Image.open('/content/Scissors.png')
    else:
        return "Invalid choice"

    display(image)

def play_game():
    user_choice = get_user_choice()
    computer_choice = get_computer_choice()
    print(f"You chose: {user_choice}")
    show(user_choice)
    print(f"Computer chose: {computer_choice}")
    show(computer_choice)
    result = determine_winner(user_choice, computer_choice)
    print(result)

```

```python
play_game()
```
